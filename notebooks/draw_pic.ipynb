```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
labels = ["RN50", "RN101", "RN50x4", "ViT-B/32"]
x = np.arange(len(labels))
width = 0.2
```

```python
#Food data
sw1 = [35.77, 36.39, 34.58, 77.67]
corner1 = [1.27, 1.22, 1.15, 71.34]
rotate1 = [9.77, 9.99, 9.85, 29.81]
watermark1=[7.73, 7.72, 8.06, 72.81]

#Pets data
sw2 = [10.3,8.68,9.29,81.84]
corner2 = [1.38, 0.27, 2.08, 76.42]
rotate2 = [19.43, 14.03, 18.02, 34.14]
watermark2=[5.42, 4.39, 3.92, 21.24]

#STL10 data
sw3 = [6.97, 5.04, 8.23, 86.20]
corner3 = [2.62, 3.01, 2.91, 7.57]
rotate3 = [43.43, 36.52, 47.98, 41.53]
watermark3=[5.90, 7.03, 6.23, 87.60]
```

```python
with plt.style.context("grayscale"):
    fig,ax1 = plt.subplots(1, 1)
    rects1 = ax1.bar(x - 1.5*width, sw1, width, label="standard whitening")
    rects2 = ax1.bar(x - 0.5*width, corner1, width, label="corner trigger")
    rects3 = ax1.bar(x + 0.5*width, rotate1, width, label="rotate trigger")
    rects4 = ax1.bar(x + 1.5*width, watermark1, width, label="watermark trigger")

    ax1.set_ylabel("% T")
    ax1.set_title("Food-101")
    ax1.set_xticks(x)
    ax1.set_xticklabels(labels)
    ax1.legend()
    

#     ax1.bar_label(rects1, padding=3)
#     ax1.bar_label(rects2, padding=3)

#     fig.tight_layout()
    plt.savefig("food.pdf")  
    plt.show()
          
```
